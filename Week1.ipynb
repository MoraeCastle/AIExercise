* **언어**

 <인공지능 번역기>

 '같은 말이라도 아 다르고 어 다르다'라는 속담이 있습니다. 서로 다른 국가의 언어를 해석하는 번역기는 무조건적인 단어의 해석만이 아닌 그 문장의 분위기를 파악해서 사용자에게 출력해야 하는데요. 저는 그 번역기 중에서 가장 보편화 되어있는 **'구글 번역기'**에 대해 알아보았습니다.

 **구글 번역**은 초반에는 SYSTRAN이라는 기계 번역 회사의 엔진을 이용하다 16년도 말부터 이세돌과 관련해서 한동안 인간vs기계 라는 주제로 화제가 된 '알파고'에도 쓰인 딥러닝 기술을 번역기에 적용하는데, 이로 인해 번역 오류를 크게 줄였다고 합니다.

 여기서 딥러닝은 머신러닝의 발전된 형태로 사람이 학습할 데이터를 입력하지 않아도 스스로 학습하고 예측한다고 합니다. 이러한 모델은 인간의 신경망을 본딴 인공 신경망에서 발전한 것인데, 이러한 기술이 적용되기 전 초기 구글 번역의 경우 입력되는 문장을 단어와 구절로 구분해 이에 맞는 외국어 단어와 구절로 옮긴 뒤 합성해 문장으로 출력하는 방식이었으나 딥러닝으로 적용하여 조각 단위가 아닌 한 번에 전체 문장을 번역하고 가장 적절한 번역을 찾아내기 위해 더 넓은 문맥을 사용하며 이후 재정렬하여 적절한 문법으로 인간이 구사하는 것과 같은 내용으로 맞춤으로서 훨씬 더 자연스러운 번역이 가능했습니다.

 이번 강의 내용처럼 현재는 하드웨어 성능이 예전과는 달리 여러 작업을 충분히 진행할 수 있는 정도까지 발전되었기 때문에 이러한 딥러닝은 현재에 이르러 가장 강력한 기술이 된 것 같습니다. 딥러닝은 머신러닝과 달리 특징을 추출하고 학습하는 모든 과정을 기계가 하기 때문에 빅데이터와 같이 방대한 데이터와 연산서버같은 자원의 크기가 클수록 더욱 강력한 인공지능이 될 수 있는 것 같습니다.
 
 저 또한 아직은 파이썬이라는 언어를 전문적으로 구현하는 레벨은 아니지만 강좌에 성실히 참여하여 딥러닝처럼 기계 스스로 학습하여 예측이 가능하도록 코딩을 할 수 있는 수준이 되고자 하는 다짐을 하게 되었습니다.


* **음성**

 <음성인식 플랫폼: 가상 비서>

 사실 이 주제도 과제에서 제시한 언어 분야와 밀접한 관련이 있는 것 같습니다. 저는 음성과 관련해 갤럭시 사용자라면 한 번쯤은 사용했을 거라 생각하는 **'빅스비'**와 관련해 조사해 보았습니다.

 빅스비는 삼성전자와 비브가 개발한 딥러닝 기반의 인공지능 음성인식 서비스입니다. 비브는 아이폰의 음성인식 서비스인 시리를 개발했던 회사인데요. 현재는 삼성에 속해 있습니다. 비브가 개발한 빅스비 서버의 구조는 캡슐이라 하는 명렁들의 집합으로 이루어져 있고 검색을 통해 해당 주제에 맞는 캡슐로 이동할 수 있다고 합니다. 따라서 사용자가 빅스비에게 명령을 하면 단어별로 나눠서 정보를 분석하고 가령 날씨를 물어봤을 경우 빅스비 서버인 브레인에서 해당 날씨 캡슐로 이동하고 측정 액션을 거치고 그 값을 사용자에게 출력하게 됩니다.

 이런 음성인식 기술은 자연 언어 처리라는 인공지능 기술을 적용하는데, 사람 음성은 그 특성상 억양이나 템포 그리고 국가별로 언어도 다르기 때문에 빅스비도 다른 음성인식 서비스와 같이 딥러닝을 적용했는데요. 사용자의 데이터가 축척될수록 더 퀄리티가 될 수 있는 이 기술이 음성인식에 적용돼 사람의 음성이 인식되면 단어별로 서버에 음성 패턴 데이터를 찾아보고 가장 가능성 있는 인식 결과를 출력하게 되므로 앞으로 빅스비 뿐만 아니라 다른 음성인식을 이용한 플랫폼들의 퀄리티 또한 높아질 것 같다는 생각을 했습니다.

 인공지능에 대해 공부하면서 기계는 0과 1로 옳고 그름을 판단하기 때문에 인공지능과 데이터는 떼려야 뗄 수 없는 관계인 것 같습니다. 이런 기술을 개발하려는 저는 이 과정을 수강하면서 완성도 있는 인공지능이 되려면 개발자가 기계에게 품질 높은 데이터를 제공해야 한다고 느끼게 됐습니다.

* **자율주행**

 <자율주행 자동차>

 자율주행도 자동차, 선박 등 하위 주제가 많지만 저는 세계적으로 많은 기업이 자율주행에서도 자동차 쪽에 많은 투자를 하고 있으므로 조사를 해봤습니다.

 자율주행은 현재는 자동차 기업뿐만 아니라 구글같은 IT 기업도 개발중인 분야입니다. 그 중에서 테슬라의 경우 누적 자율주행 자동차의 총주행거리가 30억 km가 넘었다고 합니다. 인공지능의 특성상 이런 방대한 데이터가 곧 완성도로 직결된다고 봐도 무방하기 때문에, 앞으로의 자율주행 자동차의 선두는 테슬라가 이끌어 가지 않을까 생각합니다.

 따라서 대표적으로 테슬라의 자율주행 시스템에 대해 조사해보았습니다. 테슬라는 '테슬라 오토파일럿'이라는 자율주행 시스템이 있는데 최종 목표는 운전자없이 자동차 운행이 가능할 정도입니다. 아직은 이 소프트웨어가 완성되지는 않아서 자율주행단계 중 5단계인 '완전자율주행'이 가능하지 않습니다. 비록 완전까지는 아니여도 현재는 '교통 인식 크루즈 컨트롤'이라는 운행 중에 앞차의 간격을 인식하여 간격이 좁아지면 속도를 줄이거나 '오토스티어'라는 현재 차선을 인식하여 차선 유지를 하면서 운행하는 기술로 완전 자율 주행에는 가까워 보입니다.

 하지만 자율주행의 성격상 기계에 사람의 생명을 맡겨야 하므로 논란 또한 많습니다. 실제로 우버에서 테스트 중인 자율주행 자동차가 자전거를 끌고 가던 시민과 부딪히면서 자율주행으로 인한 첫 사망 사례가 되기도 했는데요. 앞으로 자율주행 소프트웨어에 적용된 인공지능은 테스트 주행을 통해 이러한 교통사고 상황에서 제동이나 다른 차선으로 번경하는 등 성공데이터를 쌓음으로서 기능 개선을 해야될 것 같다는 생각을 했습니다.

 자율주행 같이 사람의 목숨에 가까이 관여하는 인공지능은 앞으로 IT기술이 발전할수록 자율주행 뿐만 아니라 다른 소프트웨어도 인공지능으로 대체될수록 많은 논란이 있을 것 같습니다. 여기서 저나 다른 개발자들은 완성도 있는 시스템을 위해 적절한 데이터를 구성할 줄 알아야 겠다고 생각했습니다.
